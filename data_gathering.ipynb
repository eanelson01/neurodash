After inspecting the data, there are 520 pages to go through. I will loop through each page and save the data in a json file.

```python
num_pages = 521
data = []
```

```python
for i in range(382, num_pages):
    print(i)
    parameters = {'page': i,
                'size': 500}
    r = requests.get(root + endpoint, 
            params = parameters,
            headers = headers,
            verify = False)
    total_json = json.loads(r.text)
    neurons = total_json['_embedded']['neuronResources']
    data += neurons
    time.sleep(3)
    
```

```python
r = requests.get(root + endpoint, 
                params = params,
                headers = headers)
r
```

```python
dict = json.loads(r.text)
dict['objects'][0]
```

```python
mapping = []
for i in dict['objects']:
    key = i['e']['short_name']
    value = i['e']['name']
    if {'key': key, 'value': value} not in mapping:
        mapping += [{'key': key, 'value': value}]
```

### Celltypes

```python
endpoint = 'celltypes'
r = requests.get(root + endpoint, 
                headers = headers)
r
```

```python
# numerical ids for celltypes
celltypes = json.loads(r.text)
celltypes
```

```python
celltype_data = []
for i in celltypes:
    endpoint = 'celltypes/' + str(i)
    r = requests.get(root + endpoint, 
                    headers = headers)
    data = json.loads(r.text)
    celltype_data += [data]
```

```python
with open('Data/modelsdb_celltype.json', 'w') as t:
    json.dump(celltype_data, t)
```

### Models

```python
# models
endpoint = '/models'
r = requests.get(root + endpoint, 
                headers = headers)
r
models = json.loads(r.text)
len(models)
```

```python
# have model IDs, can pull all of the data for each model based on the id
# do a test case and then expand to a for loop
models_data = []
for i in models:
    endpoint = '/models/' + str(i)
    r = requests.get(root + endpoint,
                    headers = headers)
    data = json.loads(r.text)
    models_data += [data]
```

```python
with open('Data/modelsdb_models.json', 'w') as t:
    json.dump(models_data, t)
```

### Papers

For now, I am going to just pull the papers that go along with the the models. I can get these from the paper ids from the model_data.

```python
endpoint = 'papers'
r = requests.get(root + endpoint, 
                headers = headers)
r
```

```python
# lots of papers
papers = json.loads(r.text)
len(papers)
```

```python
# get the paper ids from the models, can do this by list comprehension
# paper_ids = [x['model_paper']['value'][0]['object_id'] for x in models_data[:2]]
paper_ids = []
failed = []
for x in models_data:
    try:
        id = x['model_paper']['value'][0]['object_id']
        paper_ids += [id]
    except:
        failed += [x['id']]
```

```python
failed
```

```python
# some of these models just don't have a paper attatched to them
endpoint = 'models/' + str(failed[0])
r = requests.get(root + endpoint, 
                headers = headers)
data = json.loads(r.text)
data
```

```python
papers = []
for i in paper_ids:
    endpoint = 'papers/' + str(i)
    r = requests.get(root + endpoint, 
                headers = headers)
    data = json.loads(r.text)
    papers += [data]
```

```python
with open('Data/modeldb_papers.json', 'w') as t:
    json.dump(papers, t)
```

### modelconcepts

```python
def modeldbdata(desired_endpoint):
    r = requests.get(root + desired_endpoint, 
                    headers = headers)

    ids = json.loads(r.text)

    data = []
    
    for i in ids:
        endpoint = desired_endpoint + '/' + str(i)
        r = requests.get(root + endpoint, 
                        headers = headers)
        ind = json.loads(r.text)
        data += [ind]

    with open('Data/modeldb_' + desired_endpoint + '.json', 'w') as t:
        json.dump(data, t)
```
